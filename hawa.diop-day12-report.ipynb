```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
```

### Exercise 1:
Create a simple line plot for `y = x` for `x` values ranging from 0 to 10.

```python
x=np.arange(11)
y=x
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r"Plot of $y=x$")
plt.show()
```

### Exercise 2:
Plot `y = x^2` using Matplotlib for `x` values between -10 and 10.

```python
x=np.linspace(-10,10)
y=x**2
plt.plot(x,y, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title(b"plot of $y=x^2$")
plt.show()
```

### Exercise 3:
Create a scatter plot with 50 random points where x and y values are between 0 and 10.

```python
plt.figure(figsize=(5,8), dpi=80)
n = 50
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
plt.scatter(X,Y)
```

### Exercise 4:
Plot a sine wave using Matplotlib for `x` values between 0 and 2π.

```python
import math
X = np.linspace(0, math.pi*10, 1000,endpoint=True)
S = np.sin(X)
plt.plot(X, S)

# show the plot
plt.show()
```

### Exercise 7 and  Exercise 8:
Plot two line plots on the same graph: `y1 = x^2` and `y2 = x^3` for `x` ranging from 0 to 10.

```python
x = np.arange(11)
y1=x**2
y2=x**3
plt.plot(x,y2, 'r')
plt.plot(x,y1, 'b')
plt.xlabel('x')
plt.ylabel('y')

plt.title('plot of $y1=x^2$ $y2=x^3$' )
plt.show()
```

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
help(plt.subplot)
```

```python
x= np.arange(11)
y1=x**2
y2=x**3
plt.subplot(2,1,1)
plt.plot(y1,'r')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,1,2)
plt.plot(y2, 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.plot('y2')
```

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python
x= np.linspace(0, 5, 10)
y1=x
y2=x**2
y3=x**3
plt.figure(figsize=(8,6))
plt.plot(x,y1, 'r')
plt.plot(x,y2, 'b')
plt.plot(x,y3, 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot of $y1=x$ $y2=x^2$ $y3=x^3$' )
plt.show()
```
